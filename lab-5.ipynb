```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(42)
data=np.random.normal(0,1,100)#Normal distribution
outliers=np.random.uniform(5,10,5)#Artificial distribution
```

```python
data_with_outliers=np.concatenate([data,outliers])
```

```python
df=pd.DataFrame(data_with_outliers,columns=['value'])
df
```

```python
q1=df['value'].quantile(0.25)
q3=df['value'].quantile(0.75)
ior=q3-q1
lower_bound=q1-1.5*ior
higher_bound=q3+1.5*ior
```

```python
print(lower_bound)
print(higher_bound)
```

```python
outliers_ior=df[(df['value']<lower_bound)|(df['value']>higher_bound)]
```

```python
z_scores=stats.zscore(df['value'])
outliers_zscore=df[np.abs(z_scores)>3]
```

```python
plt.figure(figsize=(5,6))
plt.boxplot(x=df['value'])
plt.title('box plot of the outliers of distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
