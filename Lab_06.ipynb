```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import sklearn as sk
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from datetime import datetime
```

```python
data={'category':['A','B','C','A','B']}
df=pd.DataFrame(data)
df

```

```python
label_encoder=LabelEncoder()
df['Label_Encoded']=label_encoder.fit_transform(df['category'])
df
```

```python
one_hot_coding=pd.get_dummies(df['category'],prefix='category')
df=pd.concat([df,one_hot_coding],axis=1)
df
```

```python
data={'Name':['Alice','BoB'],'Date_Of_Birth':['1990-05-14','1985-08-23']}
df=pd.DataFrame(data)
df
```

```python
df['Date_Of_birth']=pd.to_datetime(df['Date_Of_Birth'])
df
```

```python
current_year=datetime.now().year
df['Age']=current_year-df['Date_Of_birth'].dt.year
df
```

```python
data={'category':['A','A','B','B','C'],'Values':[10,20,15,25,30]}
df=pd.DataFrame(data)
df
```

```python
grouped=df.groupby('category').agg({'Values':['sum','mean']})
grouped
```

```python
pivot=df.pivot_table(index='category',values='Values',aggfunc=['sum','mean'])
pivot
```
